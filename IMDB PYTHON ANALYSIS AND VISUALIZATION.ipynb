## Dealing with Missing Data

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Revenue (Millions)'].mean()
```

```python
avg_revenue = np.round(df['Revenue (Millions)'].mean(), 2)
avg_metascore = np.round(df['Metascore'].mean(), 2)
```

```python
Average of Revenue:print("A",avg_revenue)
Average of Metascore:print("Average of Metascore:",avg_metascore)
```

```python
# fill all missing Data in Revenue
df['Revenue (Millions)'].fillna(avg_revenue, inplace=True)

# fill all missing Data in Metascore
df['Metascore'].fillna(avg_metascore, inplace=True)
```

```python
df.isnull().sum()
```

```python
# remove duplicates
df.drop_duplicates(inplace=True)
```

#### Numbers of Movies Per Year

```python
no_of_movies_per_year = df.groupby(df['Year'])['Title'].count().sort_values(ascending=False)
print(no_of_movies_per_year)
```

```python
sns.barplot(no_of_movies_per_year.reset_index(), x='Year', y='Title')

plt.xlabel("Year")
plt.ylabel("Count of Movies")
plt.title("Count of Movies by Year")
plt.show()
```

```python
print(f"The Year {no_of_movies_per_year.index[0]} has the highest number of movies with a count of {no_of_movies_per_year.max()}")
```

#### Does Rating Affect Revenue?

```python
df.corr()
```

```python
#find line of best fit
a, b = np.polyfit(df['Rating'], df['Revenue (Millions)'], 1)
```

```python
plt.scatter(x=df['Rating'], y=df['Revenue (Millions)'])

#add line of best fit to plot
plt.plot(df['Rating'], a*df['Rating']+b)       

plt.xlabel("Rating")
plt.ylabel("Revenue (Millions)")
plt.title("Rating Versus Revenue")
plt.show()
```

```python
def classify_rating(rating):
    if rating >= 7.5:
        return 'Excellent'
    elif rating >= 6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
df['Rating Category'] = df['Rating'].apply(classify_rating)
```
